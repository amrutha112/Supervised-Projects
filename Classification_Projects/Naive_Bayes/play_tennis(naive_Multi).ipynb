```python
outlook=['sunny','sunny','overcast','rainy','rainy','rainy',
         'overcast','sunny','sunny','rainy','sunny',
         'overcast','overcast','rainy']
temperature=['hot','hot','hot','mild','cool','cool','cool','mild',
             'cool','mild','mild','mild','hot','mild']
humidity=['high','high','high','high','normal','normal',
          'normal','high','normal','normal','normal','high','normal','high']

wind=['weak','strong','weak','weak','weak','strong','strong','weak',
      'weak','weak','strong','strong','weak','strong']

play=['no','no','yes','yes','yes','no','yes','no','yes','yes',
      'yes','yes','yes','no']

outlook

```

```python
#encoding:to covert string into numbers
#1.label encoding
#2.one hot encoding
#3.get_dummy encoding
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

#CONVERT STRING INTO NUMBER
outlook_encoder=le.fit_transform(outlook)
print(outlook_encoder)
```

```python
temp_encoder=le.fit_transform(temperature)
print(temp_encoder)
```

```python
humidity_encoder=le.fit_transform(humidity)
humidity_encoder
```

```python
wind_encoder=le.fit_transform(wind)
wind_encoder
```

```python
#zip function:if given data is not a data frame so we can use zip function then the data will became dataframe
features=list(zip(outlook_encoder,temp_encoder,humidity_encoder,wind_encoder))
features
```

```python
#model creation
  #used naive baye's algorithm
  #used 3 type classifier
  #1.gaussian:if given dataset (input data)continuous data or numerical data
  #2.multinomial:Given dataset is discreate data
  #3.Bernouli:combination of gaussian and multinomial
  
```

```python
#model creation
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(features,play)
pred=model.predict([[2,0,0,0]])
print(pred)
```
