```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/Salary.csv")
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("Experience",fontsize=15)
plt.ylabel("Salary",fontsize=15)
plt.title("Exp VS Salary",color='m',fontsize=20)
```

```python
#split training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#here we no need for normalization
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[10.5]]))
```

```python
#To view that staightline to draw the machine
plt.scatter(x_train,y_train,color='r')

plt.plot(x_test,y_pred)
```

```python
#this straightline is known as the "best fit line".
```

```python
#equation of a line
  # y=mx+c
#machine already print m(slope) and c(constant)
```

```python
#to print slope

print("slope is",model.coef_)
```

```python
#to print the constant value
print("constant is",model.intercept_)
```

```python
#predicted value ===>y_pred
#actual value===>y_test
```

#Performance Measurement

```python
from sklearn.metrics import mean_absolute_error
print("MAE is",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print("MAPE is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print("MSE is",mean_squared_error(y_test,y_pred))
```
